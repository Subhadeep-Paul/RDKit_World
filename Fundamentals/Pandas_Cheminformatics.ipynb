## Converting Datatypes
Note, Pandas thinks that the molregno column is an integer because numerical values are written. But this is just and ID of the molecule, so converting the column to a string would be a better choice for this case. Let's fix it.

```python
df["molregno"] = df["molregno"].astype(str)
df.dtypes
```
